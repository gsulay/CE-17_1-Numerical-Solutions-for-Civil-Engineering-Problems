```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

step = 0.5
x = np.arange(0, 24+step, step)
y = (1/300)*x**3 - 0.12*x*2 + 1.94*x + np.random.normal(0, 1, len(x))
plt.plot(x, y)
plt.xlim(0, 24)
plt.xticks(np.arange(0, 25, 2))
plt.grid()
```

```python
df = pd.DataFrame({'Time (h)': x, 'Settlement (hrs)': y})
df.to_excel('Lesson 5/Numerical Diff DataGen.xlsx', index=False)
```
